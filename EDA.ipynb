```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings 
```

```python
# Load data 

df2019 = pd.read_excel('Data/Datathon_EPA_Air_Quality_Demographics_Meteorology_2019.xlsx') 
df2020 = pd.read_excel('Data/Datathon_EPA_Air_Quality_Demographics_Meteorology_2020.xlsx') 

# check the size of train and test sets
print("Number of samples in df2019 are", df2019.shape)
print("Number of samples in df2020 are", df2020.shape)
```

```python
# Check how many locations are in the sample

print(f'df2019 has {len(df2019.AQS_ID.unique())} locations.' )
print(df2019.AQS_ID.value_counts())
print('')
print(f'df2020 has {len(df2020.AQS_ID.unique())} locations.' )
print(df2020.AQS_ID.value_counts())
```

```python
# Check how many observations in each location 

# create data frame 
id19 = pd.DataFrame(df2019.AQS_ID.value_counts())
id20 = pd.DataFrame(df2019.AQS_ID.value_counts())

# then report the distribution of the number of observation 
plt.subplot(211)
plt.hist(id19.AQS_ID)
plt.title('df2019')
plt.subplot(212)
plt.hist(id20.AQS_ID)
plt.title('df2020')
plt.;

```
